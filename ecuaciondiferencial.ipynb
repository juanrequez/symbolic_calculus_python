```python
import sympy
#sympy.init_printing()
```

```python
t=sympy.Symbol("t")
```

```python
u=sympy.Function('u')
```

```python
problema1=sympy.Eq(u(t).diff(t,t)+0.125*u(t).diff(t)+u(t),0)
problema1
```

```python
usol=sympy.dsolve(problema1,u(t),ics={u(0):2, u(t).diff(t).subs(t,0):0})
usol
```

```python
sympy.plot(usol.rhs,(t,0,60),line_color='red',title='Problema 1',xlabel='tiempo, s',ylabel='distancia, m',label='u(t)',legend=True)
```
